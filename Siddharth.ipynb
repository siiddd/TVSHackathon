**Data Cleaning**

```python
data = pd.read_csv('https://raw.githubusercontent.com/siiddd/TVSHackathon/main/train.csv')
```

We can delete **'Day_Process' and 'hour_Process'**

We can use **'Day_Process' and 'hour_Process'** for Data Analysis but not in our Prediction Model.

```python
data = data.loc[:, ['ID', 'Compensation','Vehicle_Status','Loan_amt','Annuity_amt','density_Ind','Job_Type','Default_hist','Prop_List','Home_Appliances','Region_Type','Credit_Inq','Education_det','Marital_Stat','Gen','Credit_Lnks','Experience','Income_Alt','Ph.No', 'Type_Loan','Dependents','Company_Rating','Region','Age','Addr_permanent','Permanent_Addr_tag','c1_rated','c2_rated','Contact_age','Product_bought','Default']]
data.head(2)
```

Let us identify the missing values

```python
data.isna().sum()
```

```python
compensation_imputation = data['Compensation'].mean()
loanamt_imputation = data['Loan_amt'].mean()
annuity_imputation = data['Annuity_amt'].mean()
densityind_imputation = data['density_Ind'].mean()
homeappliances_imputation = data['Home_Appliances'].mean()
experience_imputation = data['Experience'].mean()
age_imputation = data['Age'].mean()
c1rated_imputation = data['c1_rated'].mean()
contactage_imputation = data['Contact_age'].mean()
productbought_imputation = data['Product_bought'].mean()
```

```python
vehiclestatus_imputation = data['Vehicle_Status'].mode()
jobtype_imputation = data['Job_Type'].mode()
defaulthist_imputation = data['Default_hist'].mode()
proplist_imputation = data['Prop_List'].mode()
regiontype_imputation = data['Region_Type'].mode()
creditinq_imputation = data['Credit_Inq'].mode()
educationdet_imputation = data['Education_det'].mode()
maritalstat_imputation = data['Marital_Stat'].mode()
gen_imputation = data['Gen'].mode()
creditlnks_imputation = data['Credit_Lnks'].mode()
incomealt_imputation = data['Income_Alt'].mode()
phno_imputation = data['Ph.No'].mode()
typeloan_imputation = data['Type_Loan'].mode()
dependents_imputation = data['Dependents'].mode()
companyrating_imputation = data['Company_Rating'].mode()
region_imputation = data['Region'].mode()
addrpermanent_imputation = data['Addr_permanent'].mode()
permanentaddrtag_imputation = data['Permanent_Addr_tag'].mode()
```

```python
jobtype_imputation
```

```python
data['Compensation'] = data['Compensation'].fillna(compensation_imputation)
data['Vehicle_Status'] = data['Vehicle_Status'].fillna(vehiclestatus_imputation[0])
data['Annuity_amt'] = data['Annuity_amt'].fillna(annuity_imputation)
data['density_Ind'] = data['density_Ind'].fillna(densityind_imputation)
data['Job_Type'] = data['Job_Type'].fillna(jobtype_imputation[0])
data['Prop_List'] = data['Prop_List'].fillna(proplist_imputation[0])
data['Home_Appliances'] = data['Home_Appliances'].fillna(homeappliances_imputation)
data['Credit_Inq'] = data['Credit_Inq'].fillna(creditinq_imputation[0])
data['Experience'] = data['Experience'].fillna(experience_imputation)
data['Income_Alt'] = data['Income_Alt'].fillna(incomealt_imputation[0])
data['Company_Rating'] = data['Company_Rating'].fillna(companyrating_imputation[0])
data['c1_rated'] = data['c1_rated'].fillna(c1rated_imputation)
data['Contact_age'] = data['Contact_age'].fillna(contactage_imputation)
```

```python
data.isna().sum()
```

Let us **LabelEncode** the Categorical Variables

```python
from sklearn.preprocessing import LabelEncoder

labelencoder = LabelEncoder()

data['Vehicle_Status'] = labelencoder.fit_transform(data['Vehicle_Status'])
data['Job_Type'] = labelencoder.fit_transform(data['Job_Type'])
data['Default_hist'] = labelencoder.fit_transform(data['Default_hist'])
data['Prop_List'] = labelencoder.fit_transform(data['Prop_List'])
data['Region_Type'] = labelencoder.fit_transform(data['Region_Type'])
data['Credit_Inq'] = labelencoder.fit_transform(data['Credit_Inq'])
data['Education_det'] = labelencoder.fit_transform(data['Education_det'])
data['Marital_Stat'] = labelencoder.fit_transform(data['Marital_Stat'])
data['Gen'] = labelencoder.fit_transform(data['Gen'])
data['Credit_Lnks'] = labelencoder.fit_transform(data['Credit_Lnks'])
data['Income_Alt'] = labelencoder.fit_transform(data['Income_Alt'])
data['Ph.No'] = labelencoder.fit_transform(data['Ph.No'])
data['Type_Loan'] = labelencoder.fit_transform(data['Type_Loan'])
data['Dependents'] = labelencoder.fit_transform(data['Dependents'])
data['Company_Rating'] = labelencoder.fit_transform(data['Company_Rating'])
data['Region'] = labelencoder.fit_transform(data['Region'])
data['Addr_permanent'] = labelencoder.fit_transform(data['Addr_permanent'])
data['Permanent_Addr_tag'] = labelencoder.fit_transform(data['Permanent_Addr_tag'])
data['Default'] = labelencoder.fit_transform(data['Default'])

data.head(2)
```

Let us check for Collinearity in the Features

```python
data.loc[:, ['Compensation', 'Loan_amt', 'Annuity_amt', 'density_Ind', 'Home_Appliances', 'Experience', 'Age', 'c1_rated', 'c2_rated', 'Contact_age', 'Product_bought']].corr()
```

```python
data = data.drop(columns = 'c2_rated', axis = 1)
```

Let us check for Data Imbalance

```python
data['Default'].value_counts()
```

Let us Handle the Imbalance 

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE()

x = data.iloc[:, :-1]
y = data.iloc[:, -1]
y = pd.DataFrame(data = y)

x_SMOTE, y_SMOTE = smote.fit_resample(x, y)

x_SMOTE['Default'] = y_SMOTE

data = x_SMOTE

data['Default'].value_counts()
```

Let us pass our data through Classification models to check performance

```python
from sklearn.model_selection import train_test_split

x = data.iloc[:, :-1]
y = data.iloc[:, -1]
y = pd.DataFrame(data = y)


x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, stratify = y)
```
