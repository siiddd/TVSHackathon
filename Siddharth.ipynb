Importing Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

**Data Cleaning**

```python
data = pd.read_csv('https://raw.githubusercontent.com/siiddd/TVSHackathon/main/train.csv')
```

```python
data.shape
```

```python
data.head(3)
```

We can delete **'Day_Process' and 'hour_Process'**

We can use **'Day_Process' and 'hour_Process'** for Data Analysis but not in our Prediction Model.

```python
data = data.loc[:, ['ID', 'Compensation','Vehicle_Status','Loan_amt','Annuity_amt','density_Ind','Job_Type','Default_hist','Prop_List','Home_Appliances','Region_Type','Credit_Inq','Education_det','Marital_Stat','Gen','Credit_Lnks','Experience','Income_Alt','Ph.No', 'Type_Loan','Dependents','Company_Rating','Region','Age','Addr_permanent','Permanent_Addr_tag','c1_rated','c2_rated','Contact_age','Product_bought','Default']]
data.head(2)
```

Let us identify the missing values

```python
data.isna().sum()
```

```python
data = data.dropna()
```

Let us **LabelEncode** the Categorical Variables

```python
from sklearn.preprocessing import LabelEncoder

labelencoder = LabelEncoder()

data['Vehicle_Status'] = labelencoder.fit_transform(data['Vehicle_Status'])
data['Job_Type'] = labelencoder.fit_transform(data['Job_Type'])
data['Default_hist'] = labelencoder.fit_transform(data['Default_hist'])
data['Prop_List'] = labelencoder.fit_transform(data['Prop_List'])
data['Region_Type'] = labelencoder.fit_transform(data['Region_Type'])
data['Credit_Inq'] = labelencoder.fit_transform(data['Credit_Inq'])
data['Education_det'] = labelencoder.fit_transform(data['Education_det'])
data['Marital_Stat'] = labelencoder.fit_transform(data['Marital_Stat'])
data['Gen'] = labelencoder.fit_transform(data['Gen'])
data['Credit_Lnks'] = labelencoder.fit_transform(data['Credit_Lnks'])
data['Income_Alt'] = labelencoder.fit_transform(data['Income_Alt'])
data['Ph.No'] = labelencoder.fit_transform(data['Ph.No'])
data['Type_Loan'] = labelencoder.fit_transform(data['Type_Loan'])
data['Dependents'] = labelencoder.fit_transform(data['Dependents'])
data['Company_Rating'] = labelencoder.fit_transform(data['Company_Rating'])
data['Region'] = labelencoder.fit_transform(data['Region'])
data['Addr_permanent'] = labelencoder.fit_transform(data['Addr_permanent'])
data['Permanent_Addr_tag'] = labelencoder.fit_transform(data['Permanent_Addr_tag'])
data['Default'] = labelencoder.fit_transform(data['Default'])

data.head(2)
```

Let us check for Collinearity in the Features

```python
data.loc[:, ['Compensation', 'Loan_amt', 'Annuity_amt', 'density_Ind', 'Home_Appliances', 'Experience', 'Age', 'c1_rated', 'c2_rated', 'Contact_age', 'Product_bought']].corr()
```

```python
data = data.drop(columns = 'c2_rated', axis = 1)
```

Let us check for Data Imbalance

```python
data['Default'].value_counts()
```

Let us Handle the Imbalance 

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE()

x = data.iloc[:, :-1]
y = data.iloc[:, -1]
y = pd.DataFrame(data = y)

x_SMOTE, y_SMOTE = smote.fit_resample(x, y)

x_SMOTE['Default'] = y_SMOTE

data = x_SMOTE

data['Default'].value_counts()
```

Let us pass our data through Classification models to check performance

```python
from sklearn.model_selection import train_test_split

x = data.iloc[:, :-1]
y = data.iloc[:, -1]
y = pd.DataFrame(data = y)


x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, stratify = y)
```

```python
y_test.value_counts()
```

# **Naive Bayes**

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()

model_NB = gnb.fit(x_train.iloc[:, 1:-1], y_train)
```

```python
y_pred = model_NB.predict(x_test.iloc[:, 1:-1])
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import classification_report
classification_report(y_test, y_pred)
```
